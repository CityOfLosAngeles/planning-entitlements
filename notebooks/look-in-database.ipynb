## tLA_PROP

* There are also APNs that are the same, but linked to different PROP_IDs.

```python
la_prop = catalog.pcts.tLA_PROP.read()
```

```python
la_prop.head()
```

```python
la_prop[la_prop.PROP_ID == 59547928]
```
